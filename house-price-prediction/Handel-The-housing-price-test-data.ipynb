```python
#IMPORT LIBRABIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# READ THE TEST DATA 

house_price_test_data=pd.read_csv('test.csv')
house_price_test_data.head()
```

```python
# CHECK THE SHAPE 

house_price_test_data.shape
```

```python
# CHECK THE NULL VALUES 


plt.figure(figsize=(15,10))
sns.heatmap(house_price_test_data.isnull())
```

```python
# HANDLE THE NULL VALUES :
```

```python
house_price_test_data['LotFrontage']=house_price_test_data['LotFrontage'].fillna(house_price_test_data['LotFrontage'].mean())

house_price_test_data['MSZoning']=house_price_test_data['MSZoning'].fillna(house_price_test_data['MSZoning'].mode()[0])

house_price_test_data.drop(['Alley'],axis=1,inplace=True)

house_price_test_data['BsmtCond']=house_price_test_data['BsmtCond'].fillna(house_price_test_data['BsmtCond'].mode()[0])

house_price_test_data['BsmtQual']=house_price_test_data['BsmtQual'].fillna(house_price_test_data['BsmtQual'].mode()[0])

house_price_test_data['FireplaceQu']=house_price_test_data['FireplaceQu'].fillna(house_price_test_data['FireplaceQu'].mode()[0])

house_price_test_data['GarageType']=house_price_test_data['GarageType'].fillna(house_price_test_data['GarageType'].mode()[0])

house_price_test_data.drop(['GarageYrBlt'],axis=1,inplace=True)

house_price_test_data['GarageFinish']=house_price_test_data['GarageFinish'].fillna(house_price_test_data['GarageFinish'].mode()[0])

house_price_test_data['GarageQual']=house_price_test_data['GarageQual'].fillna(house_price_test_data['GarageQual'].mode()[0])

house_price_test_data['GarageCond']=house_price_test_data['GarageCond'].fillna(house_price_test_data['GarageCond'].mode()[0])

house_price_test_data.drop(['PoolQC','Fence','MiscFeature'],axis=1,inplace=True)

house_price_test_data.drop(['Id'],axis=1,inplace=True)

house_price_test_data['MasVnrType']=house_price_test_data['MasVnrType'].fillna(house_price_test_data['MasVnrType'].mode()[0])

house_price_test_data['MasVnrArea']=house_price_test_data['MasVnrArea'].fillna(house_price_test_data['MasVnrArea'].mode()[0])
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(house_price_test_data.isnull())
```

```python
house_price_test_data['BsmtExposure']=house_price_test_data['BsmtExposure'].fillna(house_price_test_data['BsmtExposure'].mode()[0])

house_price_test_data['BsmtFinType2']=house_price_test_data['BsmtFinType2'].fillna(house_price_test_data['BsmtFinType2'].mode()[0])
```
