```python
plt.figure(figsize=(15,10))
sns.heatmap(house_price_training_data.isnull(),cmap='viridis')
```

```python
# DROP THE COLUMNS WHICH HAS NULL VALUES MORE THAN 50%

house_price_training_data.drop(['PoolQC','Fence','MiscFeature','Alley','GarageYrBlt'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(house_price_training_data.isnull(),cmap='viridis')
```

```python
# DROP THE ID 

house_price_training_data.drop(['Id'],axis=1,inplace=True)
```

```python
# FILL THE MISSING VALUES :

house_price_training_data['LotFrontage']=house_price_training_data['LotFrontage'].fillna(house_price_training_data['LotFrontage'].mean())
house_price_training_data['BsmtCond']=house_price_training_data['BsmtCond'].fillna(house_price_training_data['BsmtCond'].mode()[0])
house_price_training_data['BsmtQual']=house_price_training_data['BsmtQual'].fillna(house_price_training_data['BsmtQual'].mode()[0])
house_price_training_data['FireplaceQu']=house_price_training_data['FireplaceQu'].fillna(house_price_training_data['FireplaceQu'].mode()[0])
house_price_training_data['GarageType']=house_price_training_data['GarageType'].fillna(house_price_training_data['GarageType'].mode()[0])
house_price_training_data['GarageFinish']=house_price_training_data['GarageFinish'].fillna(house_price_training_data['GarageFinish'].mode()[0])
house_price_training_data['GarageQual']=house_price_training_data['GarageQual'].fillna(house_price_training_data['GarageQual'].mode()[0])
house_price_training_data['GarageQual']=house_price_training_data['GarageQual'].fillna(house_price_training_data['GarageQual'].mode()[0])
house_price_training_data['GarageQual']=house_price_training_data['GarageQual'].fillna(house_price_training_data['GarageQual'].mode()[0])
house_price_training_data['GarageCond']=house_price_training_data['GarageCond'].fillna(house_price_training_data['GarageCond'].mode()[0])
house_price_training_data['MasVnrType']=house_price_training_data['MasVnrType'].fillna(house_price_training_data['MasVnrType'].mode()[0])
house_price_training_data['MasVnrArea']=house_price_training_data['MasVnrArea'].fillna(house_price_training_data['MasVnrArea'].mode()[0])
house_price_training_data['BsmtExposure']=house_price_training_data['BsmtExposure'].fillna(house_price_training_data['BsmtExposure'].mode()[0])
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(house_price_training_data.isnull(),cmap='viridis')
```

```python
# DROP THE NULL VALUES :

house_price_training_data.dropna(inplace=True)
```

```python
# COMBINE THE TEST DATA 


house_price_prediction_test_data=pd.read_csv('formulated_house_price_test_data.csv')
house_price_prediction_test_data.head()
```

```python
final_df=pd.concat([house_price_training_data,house_price_prediction_test_data],axis=0)
final_df.head()
```

```python
# CHECK THE SHAPE OF COMBINED DATA :

final_df.shape
```

```python
# USE LABEL ENCODER TO HANDEL CATEGORICAL DATAS 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
final_df.info()
```

```python
# GET CATEGORICAL DATA
```

```python
categorical_columns=['MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood',
         'Condition2','BldgType','Condition1','HouseStyle','SaleType',
        'SaleCondition','ExterCond',
         'ExterQual','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
        'RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','Heating','HeatingQC',
         'CentralAir',
         'Electrical','KitchenQual','Functional',
         'FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive']
```

```python
final_df[categorical_columns]
```

```python
for col in categorical_columns :
    
    print(final_df[col].value_counts())
    print()
```

```python
encoding=LabelEncoder()
```

```python
for column in categorical_columns:
    final_df[column]=encoding.fit_transform(final_df[column])
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
# DIVIDE THE TRAINING AND TEST DATA 
```

```python
train_dataset=final_df.iloc[:1422,:]
test_dataset=final_df.iloc[1422:,:]
```

```python
train_dataset.head()
```

```python
train_dataset.shape
```

```python
test_dataset.shape
```

```python
# DROP THE TARGET COLUMN FROM TEST DATA

test_dataset.drop(['SalePrice'],axis=1,inplace=True)
```

```python
X_train=train_dataset.drop(['SalePrice'],axis=1)
y_train=train_dataset['SalePrice']
```

```python
sns.heatmap(X_train.isnull())
```

```python
# APPLY XGBOOST ML MODEL INTO THE DATASET :
```

```python
import xgboost
classifier=xgboost.XGBRegressor(max_depth=2,n_estimators=500)
classifier.fit(X_train,y_train)
```

```python
import xgboost
regressor_model=xgboost.XGBRegressor(max_depth=2,n_estimators=500)
```

```python
regressor_model.fit(X_train,y_train)
```

```python
import pickle
filename='semi-finalized-model.pkl'
```

```python
pickle.dump(classifier,open(filename,'wb'))
```

```python
y_pred=classifier.predict(test_dataset)
```

```python
# MAKE THE PREDICTION 

xgboost_predictions=pd.DataFrame(y_pred)
sample_submission_df=pd.read_csv('sample_submission.csv')
submission_dataset=pd.concat([sample_submission_df['Id'],xgboost_predictions],axis=1)
submission_dataset.columns=['Id','SalePrice']
```
